```python
import matplotlib as mpl
mpl.rcParams['figure.dpi'] = 100
#Use the pgf backend (must be set before pyplot imported)
#mpl.use('pgf')
#mpl.rcParams.update({
#    "pgf.texsystem": "pdflatex",
#    'font.family': 'serif',
#    'text.usetex': True,
#    'pgf.rcfonts': False,
#})
import matplotlib.pyplot as plt
from matplotlib.ticker import PercentFormatter
# Data from Eigenface and Fisherface
#The accuracy of Eigenface and the times it takes to calculate on ORL database
x = [3,4,5,6,7] #The number of face that be used to train in each class
y_e = [0.7875, 0.8, 0.8625, 0.8625, 0.875] #Accuracy of the eigenface algorithm
y_f = [0.825, 0.9, 0.9375, 0.9625, 0.975] #Accuracy of the fisherface elagorithm
y_te = [21.8, 56.2, 74, 95.9, 160.1] #Time it takes for eigenface algorithm to finish
y_tf = [8, 27, 46.1, 70.2, 103.7] #Time it takes for fisherface algorithm to finish
```

```python
#plotting
plt.plot(x, y_e, label='Eigenface', color='blue', linestyle='-', marker='o')
#plt.plot(x, y_f, label='Fisherface', color='green', linestyle='-', marker='x')

# Customize the plot
#plt.title('Two Different Lines')
plt.gca().yaxis.set_major_formatter(PercentFormatter(1))  # Format y-axis as percentages
# Set ticks for natural numbers
plt.xticks(range(min(x), max(x) + 1))  # X-axis ticks
plt.xlabel('Số lượng mặt trong training dataset')
plt.ylabel('Độ chính xác')
plt.legend()  # Show the legend
plt.grid(True)  # Add a grid
fig = plt.gcf()
fig.get_dpi()
fig.get_size_inches()
fig.set_size_inches(w=3.3, h=3)
fig.get_dpi()
plt.tight_layout()
plt.savefig('accuracyofe.png',dpi=fig.dpi)
plt.show()
```

```python
#plotting
plt.plot(x, y_te, label='Eigenface', color='blue', linestyle='-', marker='o')
plt.plot(x, y_tf, label='Fisherface', color='green', linestyle='-', marker='x')

# Customize the plot
#plt.title('Two Different Lines')
# Set ticks for natural numbers
plt.xticks(range(min(x), max(x) + 1))  # X-axis ticks
plt.xlabel('Số lượng mặt trong training dataset')
plt.ylabel('Tổng thời gian cần để phân loại (s)')
plt.legend()  # Show the legend
plt.grid(True)  # Add a grid
fig = plt.gcf()
fig.get_dpi()
fig.get_size_inches()
fig.set_size_inches(w=3.1, h=3)
fig.get_dpi()
plt.tight_layout()
plt.savefig('time.png',dpi=fig.dpi)
plt.show()
```
